```python
import osmnx as ox
import geopandas as gpd

from shapely import geometry
```

```python
zones = gpd.read_file(r'../data/hull.geojson')
hull = zones['geometry'].values[0].simplify(10)
hull
```

# osmnx filter

```python
ox.get_osm_filter('drive')
custom_filter = '["area"!~"yes"]["highway"!~"residential|cycleway|footway|path|pedestrian|steps|track|corridor|elevator|escalator|proposed|construction|bridleway|abandoned|platform|raceway|service"]["motor_vehicle"!~"no"]["motorcar"!~"no"]["access"!~"private"]["service"!~"parking|parking_aisle|driveway|private|emergency_access"]'
```

```python
ox.graph_from_place('Piedmont, California, USA', network_type='drive')
```

```python
import time

a = time.time()
custom = ox.graph_from_polygon(hull, custom_filter=custom_filter)
b = time.time()
print(b-a)
```
